### % 포맷팅 
###### C 언어 스타일의 문자열 포맷 방식, f-string이 권장되지만 여전히 레거시 코드와 일부 환경에서 사용되고 있음

#### 정렬과 폭 지정


```python
print("%s" % "Minjeong An")
print("%20s" % "Minjeong An")
print("%-20s" % "Minjeong An")
```

```python
print("%d" % 255)
print("%o" % 255)
print("%x" % 255)
print("%X" % 255)
```

```python
# 10진 정수 포맷팅

print("%10d" % 255)
print("%010d" % 255)
print("%-10d" % 255)
```

```python
# 실수 포맷팅
pi = 3.14159265

print("%f" % pi)      # 소숫점 이하 6자리
print("%.2f" % pi)    
print("%.4f" % pi)    
print("%10.3f" % pi)
print("%010.3f" % pi)
```

```python
print("성공률: %d%%" % 90)
print("C:\\Program Files\\Adobe")
print("나는 \'사과\'를 좋아합니다.")
print("나는 \"사과\"를 좋아합니다.")
```

```python
# 여러 값 출력
name = "Minjeong"
age = 25

print("이름: %s, 나이: %d" % (name, age))
```

```python
# 딕셔너리 포맷팅
person = {"name": "Minjeong", "age": 25}

print("이름: %(name)s, 나이: %(age)d" % person)
```

```python
# 표 형태 출력
students = [("Minjeong", 90), ("Alice", 85), ("Mary", 95)]

print("%-10s %3s" % ("이름", "점수"))
print("-" * 16)

for name, score in students:
    print("%-10s %5d" % (name, score))
```

#### str.format() <br>
###### 문자열을 동적으로 구성하는 유연한 출력 포맷팅 방법

```python
print("Hello, {}!".format("Minjeong"))
```

```python
# 여러 값 출력
name = "Minjeong"
age = 25

print("이름: {}".format(name))
print("이름: {}, 나이: {}".format(name, age))
```

```python
# 인덱스 지정
name = "Minjeong"
age = 25

print("이름: {0}, 나이: {1}".format(name, age))
print("나이: {1}, 이름: {0}".format(name, age))
```

```python
# 키워드 인자 사용
print("이름: {name}, 나이: {age}".format(name="Minjeong", age=25))
```

```python
# 인덱스와 키워드 혼합
print("과목: {0}, 점수: {score}".format("파이썬", score=100))
```

```python
# 정수 출력
print("{:d}".format(42))
print("{:10}".format(42))
print("{:010}".format(42))
```

```python
# 실수 출력
pi = 3.141592
print("{:.2f}".format(pi))
print("{:10.2f}".format(pi))
print("{:10.4f}".format(pi))
print("{:010.4f}".format(pi))
```

```python
# 천 단위 구분 기호
print("{:,}".format(1000000))  # 1,000,000
print("{:_}".format(1000000))  # 1,000,000
```

```python
# 퍼센트 출력
rate = 0.875
print("{:.2%}".format(rate)) 
print("{:7.1%}".format(rate))  
print("{:07.1%}".format(rate))  
```

```python
# 부호 출력
print("{:+d}".format(10))   
print("{:+d}".format(-10))  
```

```python
# 진법 변환
num = 255
print("2진수: {:b}".format(num))
print("8진수: {:o}".format(num))
print("16진수: {:x}".format(num))
print("16진수(대문자): {:X}".format(num))
```

```python
# 리스트 요소 접근
data = ["Python", 2026, 3.14]
print("언어: {0[0]}, 연도: {0[1]}, 버전: {0[2]}".format(data))
```

```python
# 딕셔너리 값 접근
person = {"name": "Minjeong", "age": 25}
print("이름: {name}, 나이: {age}".format(**person))
```

```python
# 동적 포맷팅

value = 3.141592
width = 10
precision = 2

print("{:{w}.{p}f}".format(value, w=width, p=precision))
print("{:0{w}.{p}f}".format(value, w=width, p=precision))
```

```python
# 정렬과 폭 지정
print("{:<10}".format("Python"))  # 왼쪽 정렬
print("{:>10}".format("Python"))  # 오른쪽 정렬
print("{:^10}".format("Python"))  # 가운데 정렬
```

```python
# 채우기 문자 지정
print("{:*<12}".format("Python"))
print("{:*^12}".format("Python"))
print("{:*>12}".format("Python"))
```

```python
# 날짜 및 시간 포맷팅

from datetime import datetime

now = datetime.now()
print("현재 시간: {:%Y-%m-%d %H:%M:%S}".format(now))
```

### f-string(Formatted String Literal)
###### Python 3.6 이상에서 사용할 수 있는 가장 간결하고 직관적인 문자열 포맷팅 방법



```python
name = "Minjeong"
age = 25

print(f"이름: {name}")
print(f"나이: {age}")
```

```python
# 여러 변수 출력
name = "Minjeong"
age = 25
city = "Seoul"

print(f"{name}은(는) {age}세이며 {city}에 거주합니다.")
```

```python
# 수식 포함
a = 10
b = 5

print(f"{a} + {b} = {a + b}")
print(f"{a} * {b} = {a * b}")
```

```python
# 함수 호출 결과 출력
text = "Hello~ Minjeong An~!"

print(f"대문자: {text.upper()}")
print(f"문자열 길이: {len(text)}")
```

```python
# 숫자 포맷팅
# 소수점 자리수 지정
pi = 3.14159265
print(f"{pi:.2f}")
print(f"{pi:7.2f}")
print(f"{pi:07.2f}")
print(f"{pi:10.3f}")
print(f"{pi:10.5f}")
print(f"{pi:010.5f}")
```

```python
# 천 단위 구분 기호
number = 1000000
print(f"{number:,}")
print(f"{number:_}")
```

```python
# 퍼센트 표현
rate = 0.87546
print(f"{rate:.2%}")  
print(f"{rate:.3%}")  
```

```python
# 정렬과 폭 지정
text = "Hello~ Python!"

print(f"{text:<20}")   # 왼쪽 정렬
print(f"{text:>20}")   # 오른쪽 정렬
print(f"{text:^20}")   # 가운데 정렬
```

```python
# 채우기 문자 지정
print(f"{text:*<20}")   # 왼쪽 정렬
print(f"{text:*>20}")   # 오른쪽 정렬
print(f"{text:*^20}")   # 가운데 정렬
```

```python
# 부호 및 0 채우기
num1 = 324
num2 = -324
print(f"{num1:+}")  
print(f"{num1:5}") 
print(f"{num1:05}")  
print(f"{num2:+}")  
print(f"{num2:5}") 
print(f"{num2:05}")  
```

```python
# 진법 표현
num = 124

print(f"2진수: {num:b}")
print(f"8진수: {num:o}")
print(f"16진수: {num:x}")
print(f"16진수(대문자): {num:X}")
```

```python
# 날짜와 시간 포맷팅
from datetime import datetime

now = datetime.now()
print(f"현재 시간: {now:%Y-%m-%d %H:%M:%S}")
```

```python
# 객체 속성 출력
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

s = Student("Minjeong", 100)
print(f"이름: {s.name}, 점수: {s.score}")
```

```python
# 조건문 포함
score = 85
result = "합격" if score >= 60 else "불합격"
print(f"시험 결과: {result}")

# 직접 표현식
print(f"시험 결과: {'합격' if score >= 60 else '불합격'}")
```

```python
# 디버깅 기능 (= 사용) 
# Python 3.8 이상에서 지원

x = 10 ; y = 20

print(f"{x}, {y}, {x+y}")
print(f"{x=}, {y=}, {x+y=}")
```

```python
# 여러 줄 문자열
name = "Minjeong"
score = 95.678

message = f"""
이름 : {name}
점수 : {score:.2f}
"""
print(message)
```

```python
# 종합실습
from datetime import datetime

name = "Minjeong"
score = 99.189
now = datetime.now()

message = """
이름        : {name}
점수        : {score:.2f}
합격 여부   : {result}
현재 시간   : {time:%Y-%m-%d %H:%M:%S}
""".format(
    name=name,
    score=score,
    result="합격" if score >= 60 else "불합격",
    time=now
)

print(message)
```
