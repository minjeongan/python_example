## List Comprehension
리스트 컴프리헨션 : 기존 리스트를 기반으로 새로운 리스트를 간단하게 만들어내는 방법으로, 간결하면서도 가독성이 높은 코드를 작성하는데 도움이 됨



---

```python
# 1부터 10까지의 숫자 생성
numbers = []
for n in range(1, 10+1):
    numbers.append(n)
    
numbers
```

```python
# 1부터 10까지의 숫자 생성
[n for n in range(1, 10+1)]
```

```python
squares = [x**2 for x in range(5)]
print(squares)  # [0, 1, 4, 9, 16]
```

```python
# 0부터 9까지의 숫자 중에서 짝수만을 포함하는 리스트를 생성
numbers = []
for n in range(10):
    if n % 2 == 0 : numbers.append(n)
    
numbers
```

```python
# 0부터 9까지의 숫자 중에서 짝수만을 포함하는 리스트를 생성
[n for n in range(10) if n % 2 == 0]
```

```python
# 다차원 리스트 생성 : 반복문
n_i = []
for i in range(3):
    n_j = []
    for j in range(3):
        n_j.append(i*j)
    print(n_j)       
    n_i.append(n_j) 
n_i
```

```python
# 다차원 리스트 생성 : 중첩된 리스트 컴프리헨션
[[i*j for j in range(3)] for i in range(3)]
```

### 조건식(삼항 연산자) 활용

```python
# 짝수는 그대로, 홀수는 제곱
result = [x if x % 2 == 0 else x**2 for x in range(10)]
print(result)
```

### 문자열 처리

```python
# 문자열을 대문자로 변환
words = ["python", "java", "c++"]
upper_words = [word.upper() for word in words]
print(upper_words)
```

```python
# 문자열 길이 계산
words = ["python", "java", "c++"]
lengths = [len(word) for word in words]
print(lengths)
```

```python
# 모음 제거
text = "Hello Python"
filtered = [char for char in text if char.lower() not in "aeiou"]
print(filtered)
print("".join(filtered))
```

### 중첩 반복문

```python
# 2차원 좌표 생성
pairs = [(x, y) for x in range(2) for y in range(3)]
print(pairs)
```

```python
# 3x3 행렬 생성
matrix = [[0 for _ in range(3)] for _ in range(3)]
print(matrix)
```

```python
# 리스트 평탄화(Flatten)
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flattened = [num for row in matrix for num in row]
print(flattened)
```

### 집합(Set)과 함께 중복 제거

```python
numbers = [1, 2, 2, 3, 3, 4]
unique = [x for x in set(numbers)]
print(unique)
```

### 딕셔너리(Dictionary) 데이터 처리

```python
students = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78
}
# 80점 이상 학생
top_students = [name for name, score in students.items() if score >= 80]
print(top_students)
```

```python
# 딕셔너리 컴프리헨션
squares_dict = {x: x**2 for x in range(5)}
print(squares_dict)
```

### 함수와 함께 사용

```python
def square(x):
    return x**2

result = [square(x) for x in range(5)]
print(result)
```

```python
# map()과 비교 
numbers = [1, 2, 3, 4]

lc = [x**2 for x in numbers]
mp = list(map(lambda x: x**2, numbers))

print(lc)
print(mp)
```

### 종합 실습
---

```python
# 0보다 큰 값만 필터링
numbers = [-2, -1, 0, 1, 2, -3, -10, 9]
positives = [x for x in numbers if x > 0]
print(positives)
```

```python
# 문자열에서 숫자만 추출
text = "korea 2026~~hello~! a1b2c3d4"
digits = [int(char) for char in text if char.isdigit()]
print(text)
print(digits)
```

```python
#공백 제거
words = ["  apple", "banana  ", "  cherry  "]
cleaned = [word.strip() for word in words]
print(words)
print(cleaned)
```

```python
# 두 리스트의 합
a = [1, 2, 3]
b = [4, 5, 6]
sums = [x + y for x, y in zip(a, b)]
print(sums)
```
