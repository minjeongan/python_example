```python
class Car :
	def __init__(self, speed):	# 생성자
		self.speed = speed		# 필드
	def setSpeed(self, speed):	# 메소드
		self.speed = speed
	def getDesc(self):			# 메소드
		return "차량 =("+str(self.speed) + ")"
	def __str__(self):			# 메소드 오버라이딩 -> 다형성
		return "차량 =("+str(self.speed) + ")"


class SportsCar(Car) :	# Car의 자식 클래스	-> 상속		
	def __init__(self, speed, turbo):	# 생성자
		super().__init__(speed)			# 부모 클래스의 생성자 호출
		self.turbo=turbo				# 자식 클래스에서 추가된 필드

	def setTurbo(self, turbo):			# 자식 클래스에서 추가된 메소드
		self.turbo=turbo
	def __str__(self):	# 메소드 오버라이딩 -> 다형성
		return "차량 =("+str(self.speed) + ","+str(self.turbo) + ")"

obj = SportsCar(100, True)	# obj 이름으로 SportsCar의 객체 생성
print(obj)					# __str__ 이 호출됨
print(obj.getDesc())		# Car에 정의된 메소드 호출
obj.setTurbo(False)
print(obj)
```

```python
class Animal:
    def __init__(self, name=""):
        self.name=name
    def eat(self):
        print("동물이 먹고 있습니다. ")

class Dog(Animal):      # Animal의 자식 클래스 -> 상속	
    def __init__(self):
        super().__init__()
    def eat(self):      # 메소드 오버라이딩 -> 다형성
        print("강아지가 먹고 있습니다. ")
        
class Cat(Animal):      # Animal의 자식 클래스 -> 상속	
    def __init__(self):
        super().__init__()
    def eat(self):      # 메소드 오버라이딩 -> 다형성
        print("고양이가 먹고 있습니다. ")

d = Dog()
c = Cat()
d.eat()
c.eat()
```

```python
import math
class Shape:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def area(self):
        print("계산할 수 없음!")
    def perimeter(self):
        print("계산할 수 없음!")

class Rectangle(Shape):     # Shape의 자식 클래스 -> 상속	
    def __init__(self, x, y, w, h):
        super().__init__(x, y)
        self.w = w
        self.h = h
    def area(self):         # 메소드 오버라이딩 -> 다형성
        return self.w*self.h
    def perimeter(self):    # 메소드 오버라이딩 -> 다형성
        return 2*(self.w+self.h)
    
class Circle(Shape):        # Shape의 자식 클래스 -> 상속
    def __init__(self, x, y, r):
        super().__init__(x, y)
        self.r = r
    def area(self):         # 메소드 오버라이딩 -> 다형성
        return math.pi * self.r * self.r
    def perimeter(self):    # 메소드 오버라이딩 -> 다형성
        return 2 * math.pi * self.r

r1 = Rectangle(0, 0, 100, 200)
r2 = Circle(0,0,10)

print("사각형의 면적", r1.area())
print("사각형의 둘레", r1.perimeter())

print("원의 면적", r2.area())
print("원의 둘레", r2.perimeter())
```
