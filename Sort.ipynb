### <b>정렬(Sort)</b>
##### 순서없이 배열되어 있는 자료들을 작은 것부터 큰 순서(오름차순,ascending)로 또는 큰 것부터 작은것 순서(내림차순,descending)로 재배열 하는 것

<p><b>선택정렬(Selection Sort)</b><br>
전체 원소들 중에서 기준위치에 맞는 원소를 선택하여 자리를 교환하는 방식으로 정렬<br>
시간복잡도 : 𝑶(𝒏^𝟐)
공간복잡도 : 𝑶(𝒏)</p>

```python
# 선택정렬(Selection Sort)
def selectionSort(A):
	for last in range(len(A)-1, 0, -1):
		k = theLargest(A, last)			# A[0...last] 중 가장 큰 수 A[k] 찾아 맨 뒤로 보냄
		A[k], A[last] =  A[last], A[k]	# swap : 값 교환
		print("selection Sort %d step :" % (len(A) - last), end="")
		for number in A: print("%3d" % number, end="")
		print("")

def theLargest(A, last:int) -> int:		# A[0...last]에서 가장 큰 수의 인덱스를 리턴
	largest = 0
	for i in range(last+1):
		if A[i] > A[largest]:
			largest = i
	return largest
```

```python
def main():
    A = [69, 10, 30, 2, 16, 8, 31, 22]
    print("Selection Sort Test  :", A)
    selectionSort(A)
    print("Selection Sort Result:", A)

# 파일을 직접 실행할 때만 main()이 실행됨 : 스크립트의 진입점을 명시
if __name__ == "__main__":
    main()
```

<p><b>버블정렬(Bubble Sort)</b><br>
인접한 두 개의 원소를 비교하여 자리를 교환하는 방식<br>
시간복잡도 : 𝑶(𝒏^𝟐)
공간복잡도 : 𝑶(𝒏)</p>

```python
# 버블 정렬(Bubble Sort)
def bubbleSort(A):
    for numElements in range(len(A), 0, -1):
        print("Bubble Sort %d step :" % (len(A) - numElements), end="")
        for i in range(numElements-1):
            if A[i] > A[i+1]:
                A[i], A[i+1] = A[i+1], A[i]     # swap : 값 교환
        for number in A: print("%3d" % number, end="")
        print("")
```

```python
def main():
    A = [69, 10, 30, 2, 16, 8, 31, 22]
    print("Bubble Sort Test  :", A)
    bubbleSort(A)
    print("Bubble Sort Result:", A)

if __name__ == "__main__":
    main()
```

<p><b>퀵정렬(Quick Sort)</b><br>
전체 원소에 대해서 정렬을 수행하지 않고, 기준 값을 중심으로 왼쪽 부분 집합과 오른쪽 부분 집합으로 분할하여 정렬하는 방법
<br>
시간복잡도 : 𝑶(𝒏log2n)
공간복잡도 : 𝑶(𝒏)</p>

```python
def quickSort(A, p:int, r:int):
    if p < r:
        q = partition(A, p, r)	 # 분할
        quickSort(A, p, q-1)	 # 왼쪽 부분 리스트 정렬
        quickSort(A, q+1, r)	 # 오른쪽 부분 리스트 정렬

def partition(A, p:int, r:int) -> int:
	x = A[r]					# x: 기준 원소
	i = p-1						# i: 1구역의 끝 지점
	print("Quick Sort Step   :", end="")
	for j in range(p, r):		# j: 3구역의 시작 지점
		if A[j] < x:
			i += 1
			A[i], A[j] = A[j], A[i]  # 교환
	A[i+1], A[r] = A[r], A[i+1]	   # 기준 원소와 2구역 첫 원소 교환
	for number in A: print("%3d" % number, end="")
	print("")
	return i+1
```

```python
def main():
    A = [69, 10, 30, 2, 16, 8, 31, 22]
    print("Quicksort Test  :", A)
    quickSort(A, 0, len(A)-1)
    print("Quick Sort Result:", A)

if __name__ == "__main__":
    main()
```
