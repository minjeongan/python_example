함수의 디폴트인수

```python
def process_order(customer_name, product, address=None,	payment_method=None):
    print(f"{customer_name}님의 주문이 처리되었습니다.")
    print(f"주문 내역: {product}")    
    if address:
        print(f"배송 주소: {address}")  # 배송 주소가 제공된 경우 출력
    if payment_method:
        print(f"결제 방법: {payment_method}")   # 결제 방법이 제공된 경우 출력
```

```python
process_order("Kim", "사과", address="서울시 강남구", payment_method="신용카드")
process_order("Lee", "귤", address="서울시 노원구")
process_order("Park", "포도", payment_method="카카오페이")
process_order("Choi", "복숭아")
```

함수의 가변인수

```python
def print_book_summary(title, type, published_year,	publisher, isbn, author, *authors):
    print(f"서명 : {title}")
    print(f"자료유형 : {type}")
    print(f"출판년도 : {published_year}")
    print(f"출판사 : {publisher}")
    print(f"ISBN: {isbn}") 
    print(f"저자: {author}", end="")    
    if authors:
        print(f" 외 {len(authors)}인", end="(")  # 공동 저자가 있을 경우
        for author_each in authors:
            print(author_each, end=",")
        print("\b)")
```

```python
print_book_summary("돈의 속성","단행본", 2020, "Snowfox", "9791188331796","김승호")
print("\n")
print_book_summary("(MySQL로 배우는) 데이터베이스 개론과 실습","단행본", 2024, "한빛아카데미", "9791156640127", "박우창", "남송휘", "이현룡")
```

함수의 키워드 가변인수

```python
def total_score(**kwargs):  
    print(kwargs)

    total_score = 0
    for key, value in kwargs.items():
        print(f'{key} : {value}', end="  ")
        total_score += value
    return total_score
```

```python
print("totla_score:",total_score(kor=100, math=80, eng=90))
print("totla_score:",total_score(eng=90, kor=100, math=80))
print("totla_score:",total_score(kor=100, math=80, eng=90, sci=95))
```
